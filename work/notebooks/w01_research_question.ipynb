## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
from pathlib import Path

ROOT = Path.cwd().parent.parent

df = pd.read_csv(ROOT / "data/raw/content_refresh_anonymized.csv")

declining_with_demand = df[(df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)]
print("Declining + demand:", len(declining_with_demand), "out of", len(df), "pages")

stale_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)]
print("Stale + visible:", len(stale_visible))

print(df["trend_direction"].value_counts())
```

Out of 30,000 pages, 16,262 (about 54%) are trending down — more than the 4,388 trending up. Of those declining, 13,152 still have real search demand (100+ impressions in 90 days), which is a large enough group to make ranking them worth building over the next 7 weeks.
