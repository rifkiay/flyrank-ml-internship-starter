## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML beats a fixed rule:** A fixed rule like `days_since_last_update >= 180 AND impressions_90d >= 500` can only check one or two conditions at a time. It can't learn how many different signals interact together (e.g. word count, position, engagement, freshness, all at once) to predict decline risk. The starter results back this up: the fixed baseline rule scored 0.240 on Precision@50, while a random forest model scored 0.740 on the same metric — meaning the model correctly caught about 3x more true declining pages in its top 50 picks than the rule did.

This score feeds directly into the ranked review queue from Lane 2: pages with the highest predicted decline risk appear at the top, so a reviewer knows which pages to prioritize for a content refresh action.
